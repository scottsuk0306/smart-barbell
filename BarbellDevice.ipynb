```python
import pygame
import serial
import time
from datetime import datetime

pygame.init()
port = '/dev/cu.usbmodem2101'
music_dir = "/Users/rachel/KAIST/CS/CS350/sample-15s.mp3"
sound = pygame.mixer.Sound(music_dir)
arduino = serial.Serial(port, baudrate=115200, timeout=.1)
liftTime = []
interval = []
targetPace = 5
while True:
    time.sleep(0.05)
    data = arduino.readline()
    if data:
        liftTime.append([data, datetime.now()])
        if len(liftTime)<2:
            continue
        interval.append(liftTime[-1][1]-liftTime[-2][1])
        if int(interval[-1].seconds) > targetPace:
            interval.append(999)
            sound.play()
            print(interval[-1])
            time.sleep(20)
```
